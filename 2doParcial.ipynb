```python
import random
import pandas as pd
import numpy as np
import codecs
from typing import Tuple
```

```python
def sample_csv_p(filename:str, p:float=0.01)-> pd.DataFrame:
    return pd.read_csv(filename, on_bad_lines='skip', sep='\t', header=0, skiprows=lambda x: x>0 and random.random() > p)

doc_food = codecs.open("food/en.openfoodfacts.org.products.csv",'rU','UTF-8')
df_food = sample_csv_p(doc_food)
df_food 
```

```python
filtered_df = df_food[df_food[['product_name']].notnull().all(1)]
filtered_df
```
